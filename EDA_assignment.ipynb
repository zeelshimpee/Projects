```python
# check the missing value percentage in each variable of data set (14.12)
```

```python
def defaulter_percentage (dafr,col):
    new_df=pd.DataFrame(columns=['name','percentage of defaulter'])
    for value in dafr[col].unique():
        default_cnt=dafr[(dafr[col]==value) & (dafr.TARGET==1)].shape[0]
        total_cnt=dafr[dafr[col]==value].shape[0]
        new_df=new_df.append({'name':value,'percentage of defaulter':(default_cnt*100/total_cnt)},ignore_index=True)
    return new_df.sort_values(by='percentage of defaulter',ascending=False)
```

```python
defaulter_percentage(df_merged,'NAME_SELLER_INDUSTRY')
```

- The Defaulter rate is high in 'Auto technology', 'Connectivity' in NAME_SELLER_INDUSTRY.

```python
defaulter_percentage(df_merged,'NAME_CASH_LOAN_PURPOSE')
```

- The daulter % is high for those applicants who has not mentioned the purpose of loan.

```python
defaulter_percentage(df_merged,'NAME_PORTFOLIO')
```

- The defaulters for application for Cards are higher in rate than compare to others.

```python
defaulter_percentage(df_merged,'NAME_YIELD_GROUP')
```

- The applicants whose interest rate are high have higher chances to be dafaulter.
